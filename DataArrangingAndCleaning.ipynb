```python
import pandas as pd

file_path = '/content/Delhi (potato) dataset.xlsx'
data = pd.read_excel(file_path)
```

```python
data['Price Date'] = pd.to_datetime(data['Price Date'], format='%d %b %Y')
data = data.sort_values('Price Date')
```

```python
numeric_columns = ['Min Price (Rs./Quintal)', 'Max Price (Rs./Quintal)', 'Modal Price (Rs./Quintal)']
data_aggregated = data.groupby('Price Date')[numeric_columns].mean().reset_index()
non_numeric_columns = ['State', 'Commodity']
data_non_numeric = data[['Price Date'] + non_numeric_columns].drop_duplicates(subset='Price Date')
data_final = pd.merge(data_aggregated, data_non_numeric, on='Price Date', how='left')
```

```python
full_date_range = pd.date_range(start=data_aggregated['Price Date'].min(), end=data_aggregated['Price Date'].max())
data_reindexed = data_aggregated.set_index('Price Date').reindex(full_date_range)
data_reindexed[numeric_columns] = data_reindexed[numeric_columns].interpolate(method='linear')

data_reindexed.reset_index(inplace=True)
data_reindexed.rename(columns={'index': 'Price Date'}, inplace=True)

print(data_reindexed.head())

output_file_path = '/content/drive/MyDrive/Delhi (potato) dataset.xlsx'
data_reindexed.to_excel(output_file_path, index=False)
```
